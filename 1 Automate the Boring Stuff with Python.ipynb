Parameters and variables that are assigned in a called function are said to exist in that function’s local scope . Variables that are assigned outside all functions are said to exist in the global scope . A variable that exists in a local scope is called a local variable , while a variable that exists in the global scope is called a global variable . A variable must be one or the other; it cannot be both local and global.

```python
def hello():
    apple = 5

print(apple)
```

If you need to modify a global variable from within a function, use the global statement.

```python
def hello():
    global apple
    apple = 'localaaaaa'


apple = 'globalaaaaa'
hello
print(apple)
```

Errors can be handled with try and except statements. The code that could potentially have an error is put in a try clause. The program execution moves to the start of a following except clause if an error happens.

```python
def divideByZero(divideBy):
    try:
        return 42 / divideBy
    except ZeroDivisionError:
        print('Error: You tried to divide by zero.')


print(divideByZero(2))
print(divideByZero(12))
print(divideByZero(0))
```

```python
def divideByZero(divideBy):
    return 42 / divideBy

try:
    print(divideByZero(2))
    print(divideByZero(21))
    print(divideByZero(0))
    print(divideByZero(12))   # 这行代码不会被执行，因为上面一行代码会引发异常
except ZeroDivisionError:
    print('Error: You tried to divide by zero.')  
```

```python
# 正确做法 1：在函数内部处理异常
def divideByZero(divideBy):
    try:
        return 42 / divideBy
    except ZeroDivisionError:
        return 'Error: You tried to divide by zero.'

print(divideByZero(2))
print(divideByZero(21))
print(divideByZero(0))
print(divideByZero(12))
```

```python
# 正确做法 2：在函数内部处理异常把每一次调用单独 try
def divideByZero(divideBy):
    return 42 / divideBy

for n in [2, 21, 0, 12]:
    try:
        print(divideByZero(n))
    except ZeroDivisionError:
        print('Error: You tried to divide by zero.')
```

- How many global scopes are there in a Python program? How many local scopes?

There is **one global scope**.  
There is **one local scope per function call**.

- What happens to variables in a local scope when the function call returns?
