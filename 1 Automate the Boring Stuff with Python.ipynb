Errors can be handled with try and except statements. The code that could potentially have an error is put in a try clause. The program execution moves to the start of a following except clause if an error happens.

```python
def divideByZero(divideBy):
    try:
        return 42 / divideBy
    except ZeroDivisionError:
        print('Error: You tried to divide by zero.')


print(divideByZero(2))
print(divideByZero(12))
print(divideByZero(0))
```

```python
def divideByZero(divideBy):
    return 42 / divideBy

try:
    print(divideByZero(2))
    print(divideByZero(21))
    print(divideByZero(0))
    print(divideByZero(12))   # 这行代码不会被执行，因为上面一行代码会引发异常
except ZeroDivisionError:
    print('Error: You tried to divide by zero.')  
```

```python
# 正确做法 1：在函数内部处理异常
def divideByZero(divideBy):
    try:
        return 42 / divideBy
    except ZeroDivisionError:
        return 'Error: You tried to divide by zero.'

print(divideByZero(2))
print(divideByZero(21))
print(divideByZero(0))
print(divideByZero(12))
```

```python
# 正确做法 2：在函数内部处理异常把每一次调用单独 try
def divideByZero(divideBy):
    return 42 / divideBy

for n in [2, 21, 0, 12]:
    try:
        print(divideByZero(n))
    except ZeroDivisionError:
        print('Error: You tried to divide by zero.')
```

list.index() 用来 查找某个值在列表中第一次出现的位置（索引）。

start：起始索引（包含）, end：结束索引（不包含）,从索引 start 开始“查找”，但返回的是该元素在原列表中的真实索引，不是从 start 重新编号后的索引。

```python
lst = ['a', 'b', 'c', 'b']
# index = 0 1 2 3
# value = a b c b

print(lst.index('b'))
print(lst[2:].index('b'))   # 先切片得到 ['c', 'b']这是一个新的列表对象 再对这个新的列表对象调用index方法 'b' 出现在索引 1的位置
print(lst.index('b', 2))  # 从索引2开始查找b
print(lst.index('b', 2, 4)) # 从索引2到索引4（不包括4）查找b

# list.index(value, start, end)


```

Lists of number values or lists of strings can be sorted with the sort() method.

You can also pass True for the reverse keyword argument to have sort() sort the values in reverse order.

```python
pm = [2, 0, -1, -5, 3, 5.1, 5.08]
sortedPm = sorted(pm)
print(sortedPm)
sortedPmreverse = sorted(pm, reverse=True)
print(sortedPmreverse)
```

There are three things you should note about the sort() method. 

First, the sort() method sorts the list in place; don’t try to capture the return value by writing code like spam = spam.sort()

Second, you cannot sort lists that have both number values and string values in them, since Python doesn’t know how to compare these values.

Third, sort() uses “ASCIIbetical order” rather than actual alphabetical order for sorting strings. This means uppercase letters come before lowercase letters. Therefore, the lowercase a is sorted so that it comes after the uppercase Z .

```python
aka = ['Alice', 'bob', 'Charlie', 'david', 'eve']
aka.sort()
print(aka)
```

```python
aka = ['Alice', 'bob', 'Charlie', 'david', 'eve']
aka.sort(key=str.lower)  # 按照小写字母排序 
print(aka)
```
