## Quiz 1 – Answers with Explanations

**Passing score: 75%**

---

### 1. Which of the following defines generative AI?

**Correct answer:**  
- A type of artificial intelligence that can create new content, including images, text and music.

**Explanation:**  
Generative AI is designed to produce original outputs rather than only analyzing or classifying existing data. It can generate new text, images, audio, video, and other forms of content based on patterns learned during training.

---

### 2. Which example shows that generative AI offers more than just chatbots?

**Correct answer:**  
- Creating photorealistic images of new product prototypes based on text descriptions.

**Explanation:**  
This example demonstrates generative AI’s ability to transform text into entirely new visual content. It highlights creative and design-oriented use cases, which go far beyond conversational chatbots and simple task automation.

---

### 3. Which statements accurately describe Gemini? (Select two)

**Correct answers:**  
- Gemini is an AI assistant that can help you be more creative and productive.  
- Gemini is a generative AI model (or family of models) developed by Google.

**Explanation:**  
Gemini refers both to Google’s family of generative AI models and to AI-powered tools built on top of those models. These tools are designed to assist users with writing, reasoning, problem-solving, and creative tasks, not just answering questions.

---

### 4. Which describes a multimodal gen AI application?

**Correct answer:**  
- Using gen AI to analyze customer sentiment in video testimonials and survey data.

**Explanation:**  
A multimodal generative AI application can process and understand multiple types of input, such as text, images, audio, or video. This example involves both video and text data, making it a clear case of multimodal AI usage.


## Quiz Yourself: Foundation Model or Not

For each of the following examples, decide whether it is a **foundation model** or **not**.

---

### 1. A large language model trained on massive amounts of text that can be adapted for translation, summarization, question answering, and code generation.

**Answer:** Foundation model  
**Explanation:**  
It is trained on broad, diverse data and can be reused across many different downstream tasks.

---

### 2. A multimodal model that can understand and generate text and images and be fine-tuned for tasks such as captioning and visual search.

**Answer:** Foundation model  
**Explanation:**  
The ability to handle multiple data types and support many applications is a key feature of foundation models.

---

### 3. A linear regression model used to predict house prices based on a small set of predefined features.

**Answer:** Not a foundation model  
**Explanation:**  
This model is task-specific, relies on handcrafted features, and cannot be adapted to a wide range of tasks.

---

### 4. A logistic regression model trained to classify emails as spam or not spam.

**Answer:** Not a foundation model  
**Explanation:**  
It is designed for a single classification task and lacks general-purpose learning capability.

---

### 5. A rule-based chatbot that responds to customer questions using predefined scripts.

**Answer:** Not a foundation model  
**Explanation:**  
Rule-based systems do not learn from large-scale data and cannot be transferred to other tasks.

---

### 6. A generative AI model trained on large-scale, diverse data and used as the base for many different AI-powered applications.

**Answer:** Foundation model  
**Explanation:**  
This is the defining example of a foundation model: large-scale training, generality, and reuse across tasks.
